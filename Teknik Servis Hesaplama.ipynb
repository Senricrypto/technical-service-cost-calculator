```python
# 1) Değişkenler

mesafe_km = int(input("Lütfen şirket ile servis noktası arasındaki mesafeyi km cinsinden rakam kullanarak giriniz: "))
hiz_kmh = int(input("Lütfen yol boyunca gidilen hızın ortalamsını rakam ile giriniz: "))
her_km_basina_maliyet = 0.65 ##sabit olarak değerlendireceğim

isci_saatlik_ucret = 400
onarim_suresi = float(input("Lütfen onarım/kurulum süresini saat cinsinden ondalıklı veya tam sayı olarak giriniz: "))
analiz_suresi = float(input("Lütfen analiz süresini saat cinsinden ondalıklı veya tam sayı olarak giriniz: "))
yedek_parca_ucret = float(input("Lütfen yedek parça ücreti giriniz: "))
##ileride if yapısı ile kullanılmama ihtimali eklemeyi düşünüyorum 
## Biraz daha ilerlediğimde yedek parça için örnek bir fiyatlı liste oluşturup(Excel) direk parça adı ile içeriden bilgi çekebilirim
makine_durus_maliyet = float(input("Makinenin çalışmadığı senaryolardaki ticari kaybı giriniz: "))
##Sözleşme kapsamında eklenmiş olabilir, örnek UPS sistemleri
teknisyen_sayisi = int(input("Lütfen görevlendirilen teknisyen sayısını rakam ile giriniz: "))
harici_maliyet_orani = float(input("Lütfen ondalıklı olacak şekilde harici masraf oranını rakam ile giriniz. Örnek: %35=>0.35"))

yolculuk_suresi = mesafe_km / hiz_kmh
yolculuk_masrafi = mesafe_km * her_km_basina_maliyet
toplam_calisma_suresi = onarim_suresi + analiz_suresi
iscilik_maliyeti = isci_saatlik_ucret * teknisyen_sayisi * toplam_calisma_suresi
durus_maliyeti = toplam_calisma_suresi * makine_durus_maliyet
temel_harcama = yolculuk_masrafi + iscilik_maliyeti + yedek_parca_ucret
harici_harcama = temel_harcama * harici_maliyet_orani
toplam_servis_harcamaları = temel_harcama + harici_harcama + durus_maliyeti

print("Yolculuk Masrafı:", yolculuk_masrafi)
print("Yolculuk Süresi:", yolculuk_suresi)

print("Toplam Çalışma Süresi:", toplam_calisma_suresi)
print("İsçilik Masrafı:", iscilik_maliyeti)
print("Yedek Parça Masrafı:", yedek_parca_ucret)
print("Makine Çalışmama Masrafı:", makine_durus_maliyet)
print("Temel Harcama Masrafları:", temel_harcama)
print("Harici Harcama Masrafları:", harici_harcama)
print("Toplam Harcama Tutarı:", toplam_servis_harcamaları)
```
