```python
# 1) Değişkenler

mesafe_km = int(input("Lütfen şirket ile servis noktası arasındaki mesafeyi km cinsinden rakam kullanarak giriniz: "))
hiz_kmh = int(input("Lütfen yol boyunca gidilen hızın ortalamsını rakam ile giriniz: "))
yakit_fiyat = int(input("Lütfen güncel litelik yakıt fiyatını TL cinsinden giriniz: "))##ileride kendisi güncel fiyatı API üzerinden çektireceğim

## Araç listesi
araclar = ["Renault Master","Volkswagen Crafter","Ford Transit","Fiat Ducato","Mercedes-Benz Sprinter"]
arac_no = int(input("Lütfen araç numarasını giriniz: Not: Sıralama 0 dan başlamaktadır"))
tuketim = [7.5,8.6,7.9,7.1,9.2]
km_maliyet = (tuketim[arac_no] * yakit_fiyat) / 100


#Müşteri listesi
musteriler = ["Bayern Makine,","Koru Mobiley","ZetaUPS","GammaLab","FeriCon","AzureTech","AegisPro"]
##gerçek firma isimleri değildir, varsa bile bilinçli olarak seçilmemiştir
musteri_num = int(input("Lütfen müşteri numarası giriniz: Not:Sıralama 0 dan başlamaktadır."))
isci_saatlik_ucret = 400
onarim_suresi = float(input("Lütfen onarım/kurulum süresini saat cinsinden ondalıklı veya tam sayı olarak giriniz: "))
analiz_suresi = float(input("Lütfen analiz süresini saat cinsinden ondalıklı veya tam sayı olarak giriniz: "))
yedek_parcalar = ["Trim Motoru","Enkoder","Termostat","Sıcaklık Sensörü","Baskı Tekeri","Palet Dilimi","Trim Bıcak Seti","PLC Modülü"]
yedek_parca_ucret = [5500,1200,350,260,30,26,12000,25000]
yedek_parca_no = int(input("Lütfen yedek parça numarasını giriniz: Not:Sıralama 0 dan başlamaktadır."))
##ileride if yapısı ile kullanılmama ihtimali eklemeyi düşünüyorum 
## Biraz daha ilerlediğimde yedek parça için örnek bir fiyatlı liste oluşturup(Excel) direk parça adı ile içeriden bilgi çekebilirim
makine_durus_maliyet = float(input("Makinenin çalışmadığı senaryolardaki müşterinin saatlik ticari kaybını giriniz: "))
##Sözleşme kapsamında eklenmiş olabilir, örnek UPS sistemleri
teknisyen_sayisi = int(input("Lütfen görevlendirilen teknisyen sayısını rakam ile giriniz: "))
harici_maliyet_orani = float(input("Lütfen ondalıklı olacak şekilde harici masraf oranını rakam ile giriniz. Örnek: %35=>0.35"))

yolculuk_suresi = mesafe_km / hiz_kmh
yolculuk_masrafi = mesafe_km * km_maliyet
toplam_calisma_suresi = onarim_suresi + analiz_suresi
iscilik_maliyeti = isci_saatlik_ucret * teknisyen_sayisi * toplam_calisma_suresi
durus_maliyeti = toplam_calisma_suresi * makine_durus_maliyet
temel_harcama = yolculuk_masrafi + iscilik_maliyeti + yedek_parca_ucret[yedek_parca_no]
harici_harcama = temel_harcama * harici_maliyet_orani
toplam_servis_harcamaları = temel_harcama + harici_harcama + durus_maliyeti

print("Firma Adı:",musteriler[musteri_num])
print("Servis Aracı:",araclar[arac_no])
print("Yolculuk Masrafı:", yolculuk_masrafi,"TL")
print("Yolculuk Süresi:", yolculuk_suresi,"saat")

print("Toplam Çalışma Süresi:", toplam_calisma_suresi,"saat")
print("İsçilik Masrafı:", iscilik_maliyeti,"TL")
print("Yedek Parça Masrafı:", yedek_parca_ucret[yedek_parca_no],"TL")
print("Makine Çalışmama Masrafı:", durus_maliyet,"TL")
print("Temel Harcama Masrafları:", temel_harcama,"TL")
print("Harici Harcama Masrafları:", harici_harcama,"TL")
print("Toplam Harcama Tutarı:", toplam_servis_harcamaları,"TL")
```
